```python
import pandas as pd
import spacy
import pytextrank
from itertools import chain
```

```python
data=pd.read_csv('data.csv')
```

```python
# Group by product and concatenate reviews
combined_reviews = data.groupby("ProductName")["Reviews"].apply(lambda x: "\n".join(x)).reset_index()

len(combined_reviews)
```

```python
combined_reviews
```

```python
prod_info=pd.read_csv('final_prod_info.csv')
```

```python
prod_info=prod_info.drop(columns=['Unnamed: 0.1','Unnamed: 0'])
```

```python
# Merge the DataFrames on the 'id' column
merged_df = pd.merge(combined_reviews, prod_info, on='ProductName', how='inner')
```

```python
merged_df
```

```python
len(merged_df)
```

```python
# Remove duplicate rows based on the 'final_processed' column
df = merged_df.drop_duplicates(subset='ProductName')
```

```python
len(df)
```

```python
df
```

```python
# Concatenate ProductDescription, ProductIngredients, and Reviews into a single column
df['CombinedReviews'] = df['ProductDescription'] + '\n' + df['ProductIngredients'] + '\n' + df['Reviews']

# Display the DataFrame with the combined information
df=df[['ProductName', 'CombinedReviews']]
```

```python
df
```

```python
# store the SMS message data
text_messages = df['CombinedReviews']

# Remove punctuation
processed = text_messages.str.replace(r'[^\w\d\s]', ' ')

# Replace whitespace between terms with a single space
processed = processed.str.replace(r'\s+', ' ')

# Remove leading and trailing whitespace
processed = processed.str.replace(r'^\s+|\s+?$', '')
# change words to lower case - Hello, HELLO, hello are all the same word
processed = processed.str.lower()
```

```python
df['ProcessedReviews']=processed
df
```

```python
len(df)
```

```python
nlp = spacy.load("en_core_web_sm")
# add PyTextRank to the spaCy pipeline
nlp.max_length = 2000000
nlp.add_pipe("textrank")

# new_text=processed[0]
def spacy_textProcess(text):
    processed=[]
    doc = nlp(text)
    # examine the top-ranked phrases in the document
    for phrase in doc._.phrases[:20]:
        processed.append(phrase.text)
    return processed
```

```python
df['spacyProcessedReviews']=df['ProcessedReviews'].apply(lambda x: spacy_textProcess(x) )
```

```python
df[['ProductName','spacyProcessedReviews']]
```

```python
# Flatten the list into a single list
flattened_list = list(chain.from_iterable(list(df['spacyProcessedReviews'])))
set(flattened_list)
```

```python
import re
p1=r'.*(dry|drier|rough|dehy|flaky).*'
p2=r'.*(bump|blem|acne|blackhead).*'
p3=r'.*(age|aging|young|youth|wrinkle|retinol).*'
p4=r'.*(tone|uneven|bright|dull|fair|darker).*'
p5=r'.*comb.*'
p6=r'.*hyd.*'
p7=r'.*sensit.*'
p8=r'.*(burn|eczema|irri|itch|allerg|atopic).*'
p9=r'.*(greas|oil|sticky).*'
p10=r'.*pale.*'
p11=r'.*red.*'
p12=r'.*glow.*'
p13=r'.*(sooth|smooth|soft).*'
p14=r'.*(protect|sun).*'

p15=r'.*hypoallerge.*' 
p16=r'.*((great|nice).*smell).*' 
p17=r'.*eye.*' 
p18=r'.*makeup.*' 
p19=r'.*lip.*' 
p20=r'.*(renew|health|fresh|clean|dead).*'   
p21=r'.*hand.*' 
p22=r'.*mask.*' 
p23=r'.*scrub.*' 
p24=r'.*(wipe|wash).*' 
final_processed=[]
for list1 in df['spacyProcessedReviews']:
    key=[]
    i=','.join(list1)
        
    if (re.match(p1,i)):
        key.append('dry skin')
    if (re.match(p2,i)):
        key.append('acne skin')
    if (re.match(p3,i)):
        key.append('aging skin')
    if (re.match(p4,i)):
        key.append('tone evening skin')
    if (re.match(p5,i)):
        key.append('Combination skin')
    if (re.match(p6,i)):
        key.append('hydrate skin')
    if (re.match(p7,i)):
        key.append('sensitive skin')
    if (re.match(p8,i)):
        key.append('allergic skin')
    if (re.match(p9,i)):
        key.append('oily skin')
    if (re.match(p10,i)):
        key.append('pale skin')
    if (re.match(p11,i)):
        key.append('redness prone skin')
    if (re.match(p12,i)):
        key.append('glow skin')
    if (re.match(p13,i)):
        key.append('smoothen skin')
    if (re.match(p14,i)):
        key.append('protect skin')

    if (re.match(p15,i)):
        key.append('hypoallergenic ingredients')
    if (re.match(p16,i)):
        key.append('smells good')
    if (re.match(p17,i)):
        key.append('eye concern')
    if (re.match(p18,i)):
        key.append('makeup removal')
    if (re.match(p19,i)):
        key.append('lip balm')
    if (re.match(p20,i)):
        key.append('refresh')
    if (re.match(p21,i)):
        key.append('hand cream')
    if (re.match(p22,i)):
        key.append('mask')
    if (re.match(p23,i)):
        key.append('scrubber')
    if (re.match(p24,i)):
        key.append('face wash')
    else:
        pass
    final_processed.append(list(set(key)))
```
