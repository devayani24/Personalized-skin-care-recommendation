###  Recommendation for existing Customers

```python
import ast
def recommend_products(target_user,data,processedData):
    prod_reviewed=list(set(data[data['Username']==target_user].ProductName))
    processed=(processedData[processedData.ProductName.isin(prod_reviewed)]['finalprocessed_set'])
    processed_set = set()
    for s in processed:
        processed_set |= s
#     processed_set = (sum(processed.apply(ast.literal_eval), {}))
    
    common_counts = []
    df=processedData[['ProductName','finalprocessed_set']]
    for row_set in df['finalprocessed_set']:
        common_count = len(processed_set.intersection((row_set)))
        common_counts.append(common_count)

    # Add the common counts to the DataFrame
    df['common_items_count'] = common_counts
    return  df
```

### KNN Clustering

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Convert the set of terms to a binary matrix
vectorizer = CountVectorizer(binary=True)
X = vectorizer.fit_transform(df['finalprocessed_set'].apply(lambda x: ' '.join(x)))

# Apply K-means clustering
num_clusters = 3  # Number of clusters (you can adjust this)
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(X)

# Apply PCA for visualization
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(X.toarray())

# Plot the clustered products
plt.figure(figsize=(10, 6))
for i in range(num_clusters):
    cluster_indices = np.where(clusters == i)
    plt.scatter(reduced_features[cluster_indices, 0], reduced_features[cluster_indices, 1], label=f'Cluster {i + 1}')

plt.xlabel('PCA Dimension 1')
plt.ylabel('PCA Dimension 2')
plt.title('Products Clustering')
plt.legend()
plt.tight_layout()
plt.show()
```
